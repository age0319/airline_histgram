```python
from skimage import data
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import skimage.segmentation as seg
import skimage.filters as filters
import skimage.draw as draw
import skimage.color as color
```

```python
image = data.binary_blobs()
plt.imshow(image, cmap='gray')
```

```python
image = data.astronaut()
plt.imshow(image)
```

```python
def image_show(image, nrows=1, ncols=1, cmap='gray'):
    fig, ax = plt.subplots(nrows=nrows, ncols=ncols, figsize=(14, 14))
    ax.imshow(image, cmap='gray')
    ax.axis('off')
    return fig, ax

text = data.page()
image_show(text)
```

```python
fig, ax = plt.subplots(1, 1)
ax.hist(text.ravel(), bins=32, range=[0, 256])
ax.set_xlim(0, 256);
```

```python
# text_segmented = text > (value concluded from histogram i.e 50,70,120 )
text_segmented = text > (50)
image_show(text_segmented);
```

```python
text_segmented = text > (70)
image_show(text_segmented);
```

```python
text_segmented = text > (120)
image_show(text_segmented);
```

```python
text_threshold = filters.threshold_otsu(text)  # Hit tab with the cursor after the underscore to get all the methods.
image_show(text > text_threshold);
```

```python
text_threshold = filters.threshold_li(text)  # Hit tab with the cursor after the underscore to get all the methods.
image_show(text > text_threshold);
```

```python
text_threshold = filters.threshold_local(text,block_size=51, offset=10) 
image_show(text > text_threshold);
```
